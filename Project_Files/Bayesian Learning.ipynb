```python
#Naive implementation of Probability Function that calculates joint probability of different random variables.
def probability(attributes,dframe):
    M_total = len(dframe)
    
    def applyCondition(x):
        ans = True
        for key,value in attributes.items():
            if x[key]!=value:
                ans = False
                break
        return ans
    
    M_filtered = len(dframe[dframe.apply(applyCondition, axis=1)])
    return M_filtered/M_total


#Optimized implementation of Probability Function that calculates joint probability of different random variables.
def probabilityOptimised(input_attr,probability_stats,cols,total):
    M_total = total
    
    values = []
    for attr in cols:
        values.append(input_attr[attr])
    if len(values)==1:
        values=values[0]
    else:
        values = tuple(values)
    
    M_filtered = len(probability_stats.get(values,[]))
    return M_filtered/M_total
```

```python
def baseConverter(n,precision,base=3):
    num = ""
    while n>=1:
        num += str(n % base)
        n = int(n / base)
    num = num[::-1]
    return (precision-len(num))*'0' + num
```

```python
#Naive implementation of Mutual Information of a node using probability()
def mutualInformation(X,Y,dframe,domain=nodes_domain,threshold=zero_probability):
    I = 0
    if len(Y) == 0:
        return I
    
    total = 3 ** len(Y)
    for i in domain:
        for j in range(total):
            parents_encoded = baseConverter(j,precision=len(Y))
            parents = {}
            for k in range(len(Y)):
                parents[Y[k]] = nodes_domain[int(parents_encoded[k])]
            combined = parents.copy()
            combined[X[0]] = i
            flag = True
            
            P_xy = probability(attributes=combined,dframe=dframe)
            if P_xy == 0:
                P_xy = threshold
                flag = False
            
            P_x = probability(attributes={X[0]:i},dframe=dframe)
            if P_x == 0:
                P_x = threshold
                flag = False
                
            P_y = probability(attributes=parents,dframe=dframe)
            if P_y == 0:
                P_y = threshold
                flag = False
                
            if flag == True:
                I += P_xy * math.log2(P_xy / (P_x * P_y))
    return I


#Optimized implementation of Mutual Information of a node using probabilityOptimized()
def mutualInformationOptimized(X,Y,dframe,domain=nodes_domain,threshold=zero_probability):
    I = 0
    
    if len(Y) == 0:
        return I
    
    total = len(domain) ** len(Y)
    probability_helper = []
    probability_helper.append(dframe.groupby(X).groups)
    probability_helper.append(dframe.groupby(Y).groups)
    probability_helper.append(dframe.groupby(X+Y).groups)
    M = len(dframe)
    
    for i in domain:
        for j in range(total):
            parents_encoded = baseConverter(j,precision=len(Y))
            parents = {}
            for k in range(len(Y)):
                parents[Y[k]] = domain[int(parents_encoded[k])]
            combined = parents.copy()
            combined[X[0]] = i
            
            flag = True
            P_xy = probabilityOptimised(input_attr=combined,probability_stats=probability_helper[2],cols=X+Y,total=M)
            if P_xy == 0:
                P_xy = threshold
                flag = False
            
            P_x = probabilityOptimised(input_attr={X[0]:i},probability_stats=probability_helper[0],cols=X,total=M)
            if P_x == 0:
                P_x = threshold
                flag = False
                
            P_y = probabilityOptimised(input_attr=parents,probability_stats=probability_helper[1],cols=Y,total=M)
            if P_y == 0:
                P_y = threshold
                flag = False
            
            #print(j,"\tA:",combined[X[0]],", ",parents,P_xy,P_x,P_y,x)   
            if flag:
                I += (P_xy * math.log2(P_xy / (P_x * P_y)))
    return I
```

```python
#Function to calculate Dimensionality of a node in a Graph G
def dimOfNode(G,node,domain=len(nodes_domain)):
    return (domain-1) * (domain ** len(G.predecessors(node)))
```

```python
#Function to calculate Dimensions of a Graph or Number of Independent Parameters in a Graph
def dimensionsOfGraph(G,domain=len(nodes_domain)):
    node_list = G.nodes()
    
    dim = 0
    for node in node_list:
        dim += dimOfNode(G,node)
    return dim
```

```python
#Helper function to draw a graph as well as generate a random graph
def drawGraph(G):   
    #nx.draw_networkx(G,with_labels=True,pos=nx.circular_layout(G))
    nx.draw_networkx(G,with_labels=True)
    #plt.savefig("simple_path.png")
    plt.axis('off')
    plt.show()
    
def generateGraph(n,p = 0.2,mapping = protein_mapping):
    #G = nx.fast_gnp_random_graph(n, p, directed=True)
    G = nx.gn_graph(n)
    #random_graph = nx.gnm_random_graph(n,m,directed=True)
    #G = nx.DiGraph([(u, v) for (u, v) in G.edges() if u < v])
    G = nx.relabel_nodes(G,mapping)
    return G
```

```python
#Function to calculate Entropy of a node X
def entropy(X,dframe):
    H = 0
    M = len(dframe)
    
    probability_helper = dframe.groupby(X).groups
    
    for key,value in probability_helper.items():
        P_x = len(value)/M
        H += P_x * math.log2(P_x)

    return -H
```

```python
#Function to calculate BIC Score of a Graph G
def BICScoreGraph(G,dframe,domain=nodes_domain):
    score = 0
    
    M = len(dframe)
    
    for node in G.nodes():
        I = mutualInformationOptimized([node],G.predecessors(node),dframe)
        #H = entropy([node],dframe)
        H=0
        score += I-H
    
    score *= M
    
    score -= (math.log2(M) * dimensionsOfGraph(G)) / 10 #2
    
    return score
```

```python
#Function to calculate Family Score of a node in a Graph G
def familyScore(G,X,Y,dframe,domain=nodes_domain):
    score = 0
    M = len(dframe)
    
    I = mutualInformationOptimized(X,Y,dframe)
    #H = entropy(X,dframe)
    
    #print(I,H)
    H = 0
    score += I-H
    score *= M
    
    dim = 0
    dim += dimOfNode(G,X[0])
    '''
    for node in Y:
        dim += dimOfNode(G,node)
    '''
    
    score -= (math.log2(M) * dim) / 10 #2
    
    return score
```

```python
#Function to generate a Bayesian Network corresponding to a dataset and a random initial graph
def learnGraph(dframe,G=None,no_iterations=1000,nodes_labels = protein_mapping):
    if G == None:
        G = generateGraph(11,p=(random.randrange(2,3)/10),mapping=nodes_labels)
    X = G.copy()
    itr = 0
    while itr < no_iterations:
        itr+=1
        
        try:
            oldG.clear()
        except:
            a=1
        oldG = G.copy()
        
        max_score = -1 * sys.maxsize
        max_op = None
        max_edge = None
        result = None
        
        #Section 1: Add an edge
        non_edges = list(nx.non_edges(G))
        for edge in non_edges:
            old = familyScore(G,[edge[1]],G.predecessors(edge[1]),dframe)
            temp = G.copy()
            temp.add_edge(*edge)
            if nx.is_directed_acyclic_graph(temp):
                new = familyScore(temp,[edge[1]],temp.predecessors(edge[1]),dframe)
                diff = new - old
                if diff > max_score and diff > 0:
                    max_score = diff
                    max_op = 1
                    max_edge = edge
                    if type(result) == "networkx.classes.digraph.DiGraph":
                        result.clear()
                    result = temp.copy()
                    temp.clear()
                else:
                    temp.clear()
            else:
                temp.clear()
                
        #Section 2: Delete an edge
        current_edges = list(G.edges())
        for edge in current_edges:
            old = familyScore(G,[edge[1]],G.predecessors(edge[1]),dframe)
            t = G.copy()
            t.remove_edge(*edge)
            new = familyScore(t,[edge[1]],t.predecessors(edge[1]),dframe)
            diff = new - old
            if diff > max_score and diff > 0:
                max_score = diff
                max_op = 2
                max_edge = edge
                if type(result) == "networkx.classes.digraph.DiGraph":
                    result.clear()
                result = t.copy()
                t.clear()
                
        #Section 3: Reverse an edge
        current_edges = list(G.edges())
        for edge in current_edges:
            old_x = familyScore(G,[edge[1]],G.predecessors(edge[1]),dframe)
            old_y = familyScore(G,[edge[0]],G.predecessors(edge[0]),dframe)
            t = G.copy()
            t.remove_edge(*edge)
            t.add_edge(*edge[::-1])
            if nx.is_directed_acyclic_graph(t):
                new_x = familyScore(t,[edge[1]],t.predecessors(edge[1]),dframe)
                new_y = familyScore(t,[edge[0]],t.predecessors(edge[0]),dframe)

                diff = (new_x + new_y) - (old_x + old_y)
                if diff > max_score and diff > 0:
                    max_score = diff
                    max_op = 3
                    max_edge = edge
                    if type(result) == "networkx.classes.digraph.DiGraph":
                        result.clear()
                    result = t.copy()
                    t.clear()
                else:
                    t.clear()
            else:
                t.clear()
        
        if result != None:
            G.clear()
            G = result.copy()
            result.clear()
        
        if nx.is_isomorphic(G, oldG):
            return (X,G)
        
        op = ["","Insert","Deletion","REverse"]
        print("Iteration No. #",itr,"\tScore: ",max_score,"\tEdge: ",len(G.edges()),"\tOperation Performed: ",op[max_op],"\tEdge: ",max_edge)
    return (X,G)
```

```python
#Helper Functions to derive accuracy as well as collecting stats
def accuracy(G,expected_dependencies = resultant_dependencies):
    pred_edges = len(expected_dependencies) - len(set(expected_dependencies) - set(G.edges()))
    '''
    pred_edges = 0
    for x in G.edges():
        if (x in expected_dependencies) or (x[::-1] in expected_dependencies):
            pred_edges += 1
    '''
    expected_edges = len(expected_dependencies)
    return (pred_edges/expected_edges)*100

def accuracyWithReverse(G,expected_dependencies = resultant_dependencies):
    #pred_edges = len(expected_dependencies) - len(set(expected_dependencies) - set(G.edges()))
    pred_edges = 0
    for x in G.edges():
        if (x in expected_dependencies) or (x[::-1] in expected_dependencies):
            pred_edges += 1
    expected_edges = len(expected_dependencies)
    return (pred_edges/expected_edges)*100

def drawStats(G,expected_dependencies = resultant_dependencies):
    tp = len(set(expected_dependencies))
    predicted = 0
    reverse = 0
    not_reported = 0
    for x in G.edges():
        if x in expected_dependencies :
            predicted += 1
        elif x[::-1] in expected_dependencies:
            reverse += 1
    
    not_reported = tp - predicted - reverse
    return (predicted,reverse,not_reported)
```

```python
#Function to sample from multiple graphs generated in multiple iterations and 
#selecting the best graph based on highest BIC Score
def learnDataset(dataset,n):
    graph_result = dict()
    for num,pkl in enumerate(dataset):
        print("Dataset #"+str(num),"\t",pkl)
        with open(pkl,"rb") as file:
            data = pickle.load(file)
        
        maxx = -1 * sys.maxsize
        graph = None
        for i in range(n):
            print("\t\tIteration #"+str(i))
            x = learnGraph(data,no_iterations=50)
            score = BICScoreGraph(x[1],data)
            if score > maxx:
                maxx = score
                graph = x
        
        if graph != None:
            graph_result[num] = (graph,maxx,accuracy(graph[1]),accuracyWithReverse(graph[1]))
    
    return graph_result
```

```python
#Function to sample from multiple graphs generated in multiple iterations and 
#selecting the best graph based on a confidence score on number of counts of an edges
def learnDatasetSupportScore(dataset,n,count):
    graph_result = dict()
    for num,pkl in enumerate(dataset):
        print("Dataset #"+str(num),"\t",pkl)
        with open(pkl,"rb") as file:
            data = pickle.load(file)
        
        maxx = -1 * sys.maxsize
        graph = None
        edges = []
        for i in range(n):
            print("\t\tIteration #"+str(i))
            x = learnGraph(data,no_iterations=50)
            edges.extend(x[1].edges())
        
        edges = dict(Counter(edges))
        edge = []
        for k,v in edges.items():
            if v >= count:
                edge.append(k)
        
        G = nx.DiGraph(edge)
        graph_result[num] = (G,accuracy(G),accuracyWithReverse(G),edges)
        
    return graph_result
```
